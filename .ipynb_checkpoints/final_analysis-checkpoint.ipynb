# Plotting

### Option 1
Use `pandas.DataFrame.plot` to automatically interface with matplotlib

```python
def plot_pandas(df, title, xlabel, ylabel, loglog):
    plot = df.plot(kind="line", 
        marker='o', 
        title=title, 
        xlabel=xlabel,
        ylabel=ylabel, 
        grid=True,
        loglog=loglog
    )
    return plot
```

### Option 2
Use `matplotlib.pyplot` directly for more fine-grained control. Specifically for our case we can change the x axis scaling from base 10 to base 2.

```python
def plot_mpl_2(df, df1, title, xlabel, ylabel, label1, label2):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df,
        marker='o', label=label1
    )
    ax.plot(df1, marker='>', label=label2)
    ax.plot()
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend()
    plt.grid(True)
    plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(num_threads)
    plt.show()
    return fig, ax
```

```python
def plot_mpl(df, title, xlabel, ylabel):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df,
        marker='o',
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend()
    plt.grid(True)
    plt.yscale("log", base=10)
    plt.xscale("log", base=2)
    plt.xticks(num_threads)
    plt.show()
    return fig, ax
```

```python
# Plot MemCopy Time
for size in num_vals:
    plot_mpl_2(
        df=ctk.dataframe[size[0]].loc["cudaMemcpy_device_to_host"],
        df1 = ctk.dataframe[size[0]].loc["cudaMemcpy_host_to_device"],
        title="Memcpy times for " + str(size[0]) + " values",
        xlabel="# of Threads",
        ylabel="Time (seconds)",
        label1="cudaMemcpy_device_to_host",
        label2="cudaMemcpy_host_to_device"
    )
```
